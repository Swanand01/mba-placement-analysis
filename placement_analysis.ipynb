```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Placement_Data_Full_Class.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(0.00, inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.set()
```

```python
plt.pie(x=df['degree_t'].value_counts(), labels=df['degree_t'].unique(), autopct='%1.2f%%')
plt.title('Students\' Undergraduate degree type')
```

```python
sns.countplot(x=df['gender'])
plt.title('No. of students by gender type')
plt.xlabel('Gender')
plt.ylabel('No. of students')
```

```python
sns.countplot(x=df['specialisation'])
plt.title('No. of students by Specialisation')
plt.xlabel('Specialisation')
plt.ylabel('No. of students')
```

```python
df2 = df[df['status']=='Placed']
df3 = df[df['status']=='Not Placed']
```

```python
sizes = [148, 67]
plt.pie(sizes, labels=['Placed', 'Not Placed'], autopct='%1.2f%%')
plt.title("Students' placement status")
```

```python
plt.bar(df2['specialisation'].unique(), height=df2['specialisation'].value_counts(), label='Placed')
plt.bar(df3['specialisation'].unique(), height=df3['specialisation'].value_counts(), label='Not Placed')
plt.title('No. of students placed based on their specialisation')
plt.xlabel('Specialisation')
plt.ylabel('No. of students')
plt.legend()
```

```python
plt.bar(df2['workex'].unique(), height=df2['workex'].value_counts(), label='Placed')
plt.bar(df3['workex'].unique(), height=df3['workex'].value_counts(), label='Not Placed')
plt.title('Placement based on work exp.')
plt.xlabel('Has work experience')
plt.ylabel('No. of students')
plt.legend()
```

```python
sns.scatterplot(data=df2, x=[i for i in range(148)], y='salary', hue='workex')
plt.title('Relation between work experience and salary')
plt.xlabel('Students')
plt.ylabel('Salary')
```

```python
print('Avg. MBA percentage of placed students:', round(df2['mba_p'].mean(), 2), '%')
print('Avg. Employability test percentage of placed students:', round(df2['etest_p'].mean(), 2), '%')
print('Avg. salary:', round(df2['salary'].mean()))
print('Highest salary:', df2['salary'].max())
```
